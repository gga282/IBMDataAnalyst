```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('/Users/gga/Downloads/Medical_insurance.csv')
df=df.dropna().copy()
df.describe()
```

```python
sns.scatterplot(x='age',y='charges',data=df)
plt.title('Age vs Charges')
plt.show()
```

```python
sns.scatterplot(x='bmi',y='charges',hue='smoker',data=df)
plt.title('BMI vs Charges (colored by smoker status)')
plt.show()
```

```python
sns.boxplot(x='smoker',y='charges',data=df)
plt.title('Smoker cs Charges')
plt.show()
```

```python
df_numeric=df.select_dtypes(include=['float','int'])
corr=df_numeric.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
correlation_with_charges = corr['charges'].sort_values(ascending=False)
print(correlation_with_charges)
```
